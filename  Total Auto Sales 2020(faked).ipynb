## <span id = 'jump3'>1.Which manufacturer has the highest quantity of models in this database?</span>

```python
def most_model_manufacturer(tx):
    Audi = [i for i in tx.run('match (m:model)-[:belong]->(M:Manufacturer{name:"Audi"}) return count(m), M.name')]
    Honda = [i for i in tx.run('match (m:model)-[:belong]->(M:Manufacturer{name:"Honda"}) return count(m), M.name')]
    Hongqi = [i for i in tx.run('match (m:model)-[:belong]->(M:Manufacturer{name:"Hongqi"}) return count(m), M.name')]
    Dodge = [i for i in tx.run('match (m:model)-[:belong]->(M:Manufacturer{name:"Dodge"}) return count(m), M.name')]
    BMW = [i for i in tx.run('match (m:model)-[:belong]->(M:Manufacturer{name:"BMW"}) return count(m), M.name')]
    total = [Audi, Honda, Hongqi, Dodge, BMW]
    final = Audi
    for i in total:
        if i[0].get('count(m)') > final[0].get('count(m)'):
            final = i
        else:
            pass
    print(Audi)
    print(final[0].get('M.name'), 'is the manufacturer has the most of the models as',final[0].get('count(m)'),'models')
```

## <span id ='jump4'>2.Which model is the best deal one for each horsepower?</span>

```python
def hp_per_dollar(tx):
    print('Which model is the best deal one for each horsepower?')
    name = [i for i in tx.run('match (m:model) return m.name')]
    horsepower = [i for i in tx.run('match (m:model) return m.HP')]
    price = [i for i in tx.run('match (m:model) return m.price')]
    model = 0
    h_per_d = 10000
    for i in range(len(name)):
        h_per_d_new = price[i].get('m.price')/horsepower[i].get('m.HP')
        if h_per_d_new < h_per_d:
            h_per_d = h_per_d_new
            model = name[i].get('m.name')
        else:
            pass
    print('If you purchase', model, ", you got the best deal which is", h_per_d, "dollars per horsepower")
```

```python
hp_per_dollar(db)
```

## <span id= 'jump5'>3.The average sales for each Germany model.</span>

```python
def count_german_models(tx):
    print('The average sales for each Germany model is: ')
    res_list = [res for res in tx.run('match (m:model)-[:located]->(:Country{country:"German"}) return m.sale')]
    total = 0
    for i in range(len(res_list)):
        total += res_list[i].get('m.sale')
    print(total/len(res_list))
```

```python
count_german_models(db)
```

## <span id = 'jump6'>4.Which manufacturer have the hightest sales?</span>

```python
def print_highest_Sales(tx):
    Audi = [i for i in tx.run('match (m:model)-[:belong]->(M:Manufacturer{name:"Audi"}) return sum(m.sale), M.name')]
    Honda = [i for i in tx.run('match (m:model)-[:belong]->(M:Manufacturer{name:"Honda"}) return sum(m.sale), M.name')]
    Hongqi = [i for i in tx.run('match (m:model)-[:belong]->(M:Manufacturer{name:"Hongqi"}) return sum(m.sale), M.name')]
    Dodge = [i for i in tx.run('match (m:model)-[:belong]->(M:Manufacturer{name:"Dodge"}) return sum(m.sale), M.name')]
    BMW = [i for i in tx.run('match (m:model)-[:belong]->(M:Manufacturer{name:"BMW"}) return sum(m.sale), M.name')]
    total = [Audi, Honda, Hongqi, Dodge, BMW]
    final = Audi
    for i in total:
        if i[0].get('sum(m.sale)') > final[0].get('sum(m.sale)'):
            final = i
        else:
            pass
    print(final[0].get('M.name'), 'is the manufacturer has the highest sales of all its models as',final[0].get('sum(m.sale)'),'sales.')
```

## <span id='jump7'>5.Compare to the horsepower, which model has the best performence on MPG?</span>

```python
def mpg_per_hp(tx):
    print('Compare to the horsepower, which model has the best performence on MPG?')
    name = [i for i in tx.run('match (m:model) return m.name')]
    horsepower = [i for i in tx.run('match (m:model) return m.HP')]
    mpg = [i for i in tx.run('match (m:model) return m.MPG')]
    model = 0
    m_per_h = 0
    for i in range(len(name)):
        m_per_h_new = horsepower[i].get('m.HP')/mpg[i].get('m.MPG')
        if m_per_h_new > m_per_h:
            m_per_h = m_per_h_new
            model = name[i].get('m.name')
        else:
            pass
    print('If you purchase', model, ", you got the best deal which is", m_per_h, "horsepower for Miles Per Gallon")
```

```python
mpg_per_hp(db)
```
